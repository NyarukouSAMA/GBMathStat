### Задача 1

Подбрасываются три игральных кубика. Оцените вероятность того, что сумма значений этих кубиков не будет превышать 6.

Подсказка: оценкой вероятности может служить, например, относительная частота при большом числе опытов. Попробуйте реализовать наибольшее число опытов, которое позволяют ваш компьютер и/или ваше терпение.

Замечание: вообще в этом курсе мы не используем такие численные методы оценки вероятностей, и в будущем мы не будем их применять. Но тут сделаем исключение один раз.

Перебор возможных комбинаций для сумм. Вероятности у всех них одинаковые, все комбинации элементарные события. Собственно, тогда нужно найти количество комбинаций, в которых сумма не больше 6. По сути (прочитал в википедии) - это задача поиска способов разбить число на n слагаемые (это кстати не точно, определение я не совсем понял). Но я решил не париться с формулами из вики, и путем перебора получить количество возможных комбинаций из всех, для которых сумма меньше или равна 6. Ну а вероятность 1 конкретного вырианта найти не сложно.

```python
dice_variences = np.array([[i,j,k] for i in range(1,7) for j in range(1,7) for k in range(1,7)])
```

```python
suitable_issue = np.array([[]], dtype=int)

for dice_variant in dice_variences:
    if (dice_variant.sum() <= 6):
        if suitable_issue.size == 0:
            suitable_issue = np.append(suitable_issue, [dice_variant], axis = 1)
        elif (suitable_issue.ndim == 2 and not(suitable_issue == dice_variant).all(1).any()):
            suitable_issue = np.append(suitable_issue, [dice_variant], axis = 0)
```

```python
p_single = 1/dice_variences.shape[0]
p_for_issue = suitable_issue.shape[0] * p_single
print(p_for_issue)
```

Для сумм ниже 7 количество вариантов 20. Все они равновероятны. А т.к. вероятность 1 исхода равна ($P(single) = \dfrac{1}{n}$), где n кол-во комбинаций 3 кубиков, то посчитать можно умножив на кол-во исходов. P = 0.09

### Задача 2

Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.

Задача сводится к поиску числа комбинаций, т.к. порядок нас не интересует. Нужно вытянуть 4 карты так, что-бы среди них был хотя-бы 1 туз. Т.е. нвыми словами нас не устраивают комбинации, в которых тузов нет. Т.е. иными словами нам нужно определить число комбинаций для колоды без тузов. Все эти комбинации нас не интересуют. Тогда вычтем из всех возможных комбинаций те, которые без тузов. Это и будет ответ.

```python
def combinations(n: int, k: int) -> int:
    """Число сочетаний.
    """
    
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
```

```python
combinations(52,4) - combinations(48,4)
```

### Задача 4

Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми? Почему?

Интересная задача. Оговоримся, что у кубика 6 граней. Тогда, вероятность выпадения конкретной грани 1/6.

Событие A по условию - четное число, на кубике четных чисел 3, следовательно вероятность такого события - 3/6, или 1/2

Событие B по условию - появление числа больше 3. Это тоже 1/2

Теперь совместим события А и B - необходимо, что-бы выпавшее число было одновременно и больше 3 и четным. Таких на кубике 2 числа: 4 и 6, следовательно вероятность  выпадения четого больше 3 равна 2/6 или 1/3

Из формулы вероятности совместных независимых событий $P(AB) = P(A) * P(B)$, т.е. $1/2 * 1/2 = 1/3$, что не является корректным, потому-что $1/4 \ne 1/3$

Следовательно события A и B зависимы.

### Задача 5

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ. Какова вероятность, что вы действительно больны ей?

Подсказка: используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

Вероятность того, что вы поражены болезнью 1/1000, при этом, с вероятностью 99/100 - тест положителен, тогда, когда вы больны, и только 1/100, что тест положительный не смотря на то, что вы здоровы. Т.е. положеительный тест - это событие B, условное от события A.

Будем считать факт заболеваний событием A. $P(A) = 1/1000$
Вероятность того, что тест покажет положительный результат для больного - это B по A. $P(B|A) = 99/100$

Нам известно, что условная вероятность вычисляется по формуле $P(B|A) = \dfrac{P(BA)}{P(A)}$

Тогда вероятность того, что вы больны и тест положительный (совместная вероятность) - это $P(BA) = P(B)*P(B|A)$

```python
print((1/1000)*(99/100))
```

Однако нам пока достоверно неизвестно, действительно-и мы больны этой болезнью. Вероятность того, что мы действительно больны болезнью, при условиии, что тест положительный - это тоже условная вероятность, правда обратная - $P(A|B)$. И тут мы можем воспользоваться формулой Байса для разворота обратной вероятности.

$$P(A|B) = \dfrac{P(B|A) \cdot P(A)}{P(B)}$$

Однако P(B) - общая вероятность положительного теста нам неизвестна. С другой стороны нам известно, что ложно положительных у нас 1%, т.е. вероятность того, что тест будет положительный, в случае когда вы не больны, составляет 1%. $P(B|\overline{A}) = 1/100$. Так-же нам известна вероятность того, что мы не больны - это $P(\overline{A}) = 999/1000$

Т.е. можно вычислить полную вероятность для положительного результата теста. Она вычисляется по формуле общей вероятности. А т.к. события два, A и не A, то все имеющиеся данные у нас уже есть

```python
P_B = (99/100)*(1/1000)+(1/100)*(999/1000)
print(P_B)
```

И так:
- $P(B|A) = 99/100$
- $P(A) = 1/1000$
- $P(B)$ - записано в перемменную
Осталось вычислить искомый результат:

```python
P_AinB = (99/100)*(1/1000)/P_B
print(P_AinB)
```

Вероятность того, что вы действительно больны около 9%
