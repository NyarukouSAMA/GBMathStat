3. Пример с подсчетом полной вероятности (12-ый из practice1) не очень нагляден, если честно. Потому что там по факту идет подсчет следующим образом:
$$P(A) = \frac{5}{8} \cdot \frac{1}{3} + 0 \cdot \frac{1}{3} + 1 \cdot \frac{1}{3} = \frac{5}{24} + 0 + \frac{1}{3} = \frac{13}{24} \approx 0.5417$$

Особнно вот тут: $P(A | H_1) = \dfrac{5}{3 + 5} = \dfrac{5}{8}$

Исходя из используемого метода подсчета условной вероятности $A$ по $H_1$ получается, что события $A$ и $H_1$ независимы (что логично, вероятность выбора корзины не влияет на вероятность выбора мяча, т.к. мяч мы выбираем независимо от выбранной корзины). Но тогда получается, что условная вероятность события $A$ по $H_1$ равно вероятности наступления события $A$.

Т.е. $P(A | H_1) = P(A)$ для независимых событий. Эту формулу возможно стоило-бы указать в методичке.
