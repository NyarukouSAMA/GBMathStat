### Дополнительные материалы

__Что такое смещённая и несмещённая оценки?__

Пусть $x$ — случайная величина. Рассмотрим некоторую статистику $S(X)$, реализующую оценку какого-то параметра случайной величины $x$ (это может быть выборочное среднее $S(X) = \overline{X}$, выборочная дисперсия $S(X) = \sigma_X^2$ или ещё что-то). Мы можем взять некоторую выборку $X$ размера $n$ из случайной величины $x$ и посчитать по ней значение $S(X)$. Затем можно взять ещё одну выборку, посчитать от неё $S(X)$. И ещё, и ещё.

В результате можно получить новую выборку из всевозможных значений
$$S(X_1), \: S(X_2), \: \dots, \: S(X_m),$$
где $m$ — то, сколько раз мы брали новую выборку.

Ранее мы уже отмечали, что любая выборка представляет собой значения какой-то случайной величины. Выходит, и выборка $S(X_1)$, $S(X_2)$, ..., $S(X_m)$ взята из какого-то распределения $y$. Это распределение представляет собой всевозможные значения статистики $S(X)$ по выборкам размера $n$ из случайной величины $x$. (О таких распределениях мы ещё поговорим на уроке 4, когда будем говорить о Центральной предельной теореме.)

Итак, оценка $S(X)$ называется __несмещённой__, если математическое ожидание этой оценки равняется реальному значению этого параметра.

Например, пусть $S(X) = \overline{X}$. С помощью такой статистики, как мы уже знаем, можно оценить математическое ожидание $M(x)$ случайной величины $x$. Такая оценка действительно является несмещённой:
$$M \left( \overline{X} \right) 
    = M \left( \dfrac{1}{n} \displaystyle\sum_{i = 1}^n x_i \right) 
    = \dfrac{1}{n} \displaystyle\sum_{i = 1}^n M(x_i)
    = \dfrac{1}{n} \displaystyle\sum_{i = 1}^n M(x)
    = M(x)$$

В практическом смысле это означает, что если мы рассмотрим большое количество различных выборок, то, хотя выборочное среднее каждой из них вряд ли будет равно математическому ожиданию $x$, в среднем мы получим именно его.
    
Оказывается, обычная (смещённая) оценка дисперсии является смещённой ([источник](https://en.wikipedia.org/wiki/Variance#Sample_variance)):
$$M \left(\sigma_{X, \: biased}^2 \right) = \dfrac{n - 1}{n} D(x)$$

Поэтому на практике она не используется, а вместо неё используется её несмещённый вариант:
$$\sigma_X^2 = \dfrac{1}{n - 1} \displaystyle\sum_{i = 1}^n \left( x_i - \overline{X} \right)^2$$

__Почему именно 1?__

Почему для получения несмещённой оценки мы вычитаем именно 1? На самом деле это не совсем так, в будущем мы будем вычитать и 2, и другие значения. Общая идея тут состоит в подсчёте __числа степеней свободы__.

__Числом степеней свободы__ в статистике называют число независимых переменных за вычетом числа промежуточных оценок, использованных в процессе построения окончательной оценки. 

Например, при вычислении выборочного среднего мы не используем никакие промежуточные оценки, поэтому число степеней свободы тут равно $n$.

В процессе же построения оценки дисперсии мы используем выборочное среднее, которое само в свою очередь является оценкой. Поэтому число степеней свободы равно $n - 1$.

Например, при вычислении дисперсии ошибки модели линейной регрессии (об этом поговорим на уроке 7) используются посчитанные в процессе коэффициенты линейной регрессии, поэтому число степеней свободы там вовсе равно $n - k$, где $k$ — число коэффициентов.
